```python
import pandas as pd
import os

```

```python
tags = os.listdir("meta")
final = pd.DataFrame(columns=["Tag","Slno", "Title", "Difficulty","Time", "Space","Favorite", "Acceptance"])
def fillfav(col):
    if pd.isnull(col['Favorite']):
        return col['Favorite']
    else:
        return "yes"      
for tag in tags:
    data = pd.read_csv(f'meta/{tag}', sep="|", header=None)
    data.columns = ["Slno", "Title", "Solution", "Difficulty","Time", "Space","Favorite", "Acceptance"]
    data["Slno"] = data["Title"].str.split(".").str[0]
    data["Tag"] = [tag]*data.shape[0]
    data["Favorite"] = data[["Favorite"]].apply(fillfav,axis=1)
    data.drop(["Solution"],axis =1,inplace =True)
    data = data[["Tag","Slno", "Title", "Difficulty","Time", "Space","Favorite", "Acceptance"]]
    final = pd.concat([final,data])
```
